## Investigation Overview


> In this investigation, I wanted to look at the characteristics of GoBike that could be used to predict their duration seconds. The main focus was on the three variables of GoBike: bike_id, member_gender, and bike_share_for_all_trip.

## Dataset Overview

> This data set includes information about individual rides made in a bike-sharing system covering the greater San Francisco Bay area.
The start_time and end_time were object data type, I changed them to datetime data type and split the date and time into four variables start_time, end_time, start_date, and end_date.
I also remove all the NAN values to make the dataset ready for analysis.

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
df_bike = pd.read_csv('fordgobike-tripdata.csv')
```

## Distribution of GoBike Duration Seconds
Duration second has a long-tailed distribution, with a lot of small time end, and few on the greatest time end.
When plotted on a log-scale, the duration seconds distribution looks a normal distribution, with the peak between 20000 and 60000, and its slightly skewed to the right.

```python
# there's a long tail in the distribution, so let's put it on a log scale instead
log_binsize = 0.01
bins = 10 ** np.arange(1.0, np.log10(df_bike['duration_sec'].max())+log_binsize, log_binsize)

plt.figure(figsize = [14,8])
plt.hist(data = df_bike, x = 'duration_sec', bins = bins)
plt.xscale('log')
plt.xticks([1e1, 1e2, 1e3, 1e4, 1e5], [100, '20k', '40k', '60k', '80k'])
plt.xlabel('Time')
plt.suptitle("Distribution of GoBike Duration Seconds".title(), y =1,
            fontsize = 14, weight = "bold")
plt.show()
```

## Duration Seconds vs. member gender
>From the graph we can observe under the member gender column the male had the least duration second indcating the fasters, female being the second and the others has the least number of bike_id

```python
fig = plt.figure(figsize = [5,4])
def_color = sb.color_palette()[0]
sb.pointplot(data=df_bike, x='member_gender', y='duration_sec', color=def_color, errorbar='sd', linestyles="")
plt.xticks(rotation=15);
plt.ylabel('duration_sec');
plt.suptitle("Duration Seconds vs. member gender".title(), y =1,
            fontsize = 14, weight = "bold");
```

## Duration Seconds vs. (member gender and bike_share_for_all_trip)

We can see how member_gender and duration_sec grades affect GoBike duration second by plotting the average duration second on selected duration seconds: those around 100 duration_sec and those around 1000 duration_sec. For both the 100-duration_sec and 1000-duration_sec GoBike, as the  duration second level increases, the general height of the points rises, showing an increase in duration second.

```python
df_bike_flag = (df_bike['duration_sec'] >= 100) & (df_bike['duration_sec'] <= 1000)
df_bike_sml = df_bike.loc[df_bike_flag,:]
```

```python
fig = plt.figure(figsize = [14,8])
ax = sb.pointplot(data = df_bike_sml, x = 'member_gender', y = 'duration_sec', hue = 'bike_share_for_all_trip',
           palette = 'Blues', linestyles = '', dodge = 0.4)
plt.suptitle("Duration Seconds vs. (member gender and bike_share_for_all_trip)".title(), y =1,
            fontsize = 14, weight = "bold")
plt.ylabel('duration_sec')
plt.show();
```
