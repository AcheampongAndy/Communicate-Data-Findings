## Preliminary Wrangling


```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline
```

```python
df_bike = pd.read_csv('201902-fordgobike-tripdata.csv')
df_bike.head(10)
```

```python
df_bike.info()
```

> Splitting the date and time in the start_time and end_time columns into a different columns

```python
import datetime
df_bike['start_time'] = pd.to_datetime(df_bike['start_time'])
df_bike['end_time'] = pd.to_datetime(df_bike['end_time'])
```

```python
df_bike['stat_tim'] = [d.time() for d in df_bike['start_time']]
df_bike['end_tim'] = [d.time() for d in df_bike['end_time']]
df_bike['start_date'] = [d.date() for d in df_bike['start_time']]
df_bike['end_date'] = [d.date() for d in df_bike['end_time']]
df_bike.drop(['start_time', 'end_time'], axis = 1, inplace = True)
```

```python
df_bike.rename(columns={'stat_tim':'start_time', 'end_tim': 'end_time'}, inplace = True)
```

```python
df_bike = df_bike[['duration_sec',
 'start_time',
 'end_time',
 'start_date',
 'end_date',
 'start_station_id',
 'start_station_name',
 'start_station_latitude',
 'start_station_longitude',
 'end_station_id',
 'end_station_name',
 'end_station_latitude',
 'end_station_longitude',
 'bike_id',
 'user_type',
 'member_birth_year',
 'member_gender',
 'bike_share_for_all_trip',
 ]]
df_bike.head(10)
```

> Removing rows that contain no content

```python
df_bike = df_bike.dropna().reset_index(drop=True)
df_bike.info()
```

```python
df_bike.shape
```

```python
df_bike.dtypes
```

```python
plt.figure(figsize = [5, 10])


def_color = sb.color_palette()[0]
plt.subplot(3, 1, 1)
sb.countplot(data = df_bike, x = 'user_type', color = def_color, );

plt.subplot(3, 1, 2)
sb.countplot(data = df_bike, x = 'member_gender', color = def_color,);

plt.subplot(3, 1, 3)
sb.countplot(data = df_bike, x = 'bike_share_for_all_trip', color = def_color,);
```

```python
df_bike.describe()
```

```python
# start with a standard-scaled plot
binsize = 100
bins = np.arange(0, df_bike['duration_sec'].max()+binsize, binsize)

plt.figure(figsize=[8,5])
plt.hist(data = df_bike, x = 'duration_sec', bins = bins)
plt.xlabel('Time')
plt.show()
```

```python
# there's a long tail in the distribution, so let's put it on a log scale instead
log_binsize = 0.01
bins = 10 ** np.arange(1.0, np.log10(df_bike['duration_sec'].max())+log_binsize, log_binsize)

plt.figure(figsize = [8,5])
plt.hist(data = df_bike, x = 'duration_sec', bins = bins)
plt.xscale('log')
plt.xticks([1e1, 1e2, 1e3, 1e4, 1e5], [100, '20k', '40k', '60k', '80k'])
plt.xlabel('Time')
plt.show()
```

Duration second has a long-tailed distribution, with a lot of small time end, and few on the greatest time end.
When plotted on a log-scale, the duration seconds distribution looks a normal distribution, with the peak between 20000 and 60000, and its slightly skewed to the right.

```python
fig, ax = plt.subplots(nrows=3, figsize = [8,8])

variables = ['start_station_id', 'end_station_id', 'bike_id']
for i in range(len(variables)):
    var = variables[i]
    bins = np.arange(min(df_bike[var]), max(df_bike[var])+3, 3)
    ax[i].hist(data = df_bike, x = var, bins = bins)
    ax[i].set_xlabel('{} (mm)'.format(var))

plt.show()
```

```python
# plotting bike_id on a standard scale
binsize = 50
bins = np.arange(0, df_bike['bike_id'].max()+binsize, binsize)

plt.figure(figsize=[8, 5])
plt.hist(data = df_bike, x = 'bike_id', bins = bins)
plt.xlabel('bike_id')
plt.show()
```

## Bivariate Exploration



```python
numeric_vars = ['duration_sec','start_station_id','start_station_latitude','start_station_longitude',
                'end_station_id','end_station_latitude','end_station_longitude','bike_id','member_birth_year']
categoric_vars = ['user_type', 'member_gender','bike_share_for_all_trip']
```

```python
# correlation plot
plt.figure(figsize = [8, 5])
sb.heatmap(df_bike[numeric_vars].corr(), annot = True, fmt = '.3f',
           cmap = 'vlag_r', center = 0)
plt.show()
```

```python
# plot matrix: sample 500 diamonds so that plots are clearer and they render faster
print("df_bike.shape=",df_bike.shape)
df_bike_samp = df_bike.sample(n=500, replace = False)
print("df_bike_samp.shape=",df_bike_samp.shape)

g = sb.PairGrid(data = df_bike_samp, vars = numeric_vars)
g = g.map_diag(plt.hist, bins = 20);
g.map_offdiag(plt.scatter)
```

```python
# plot matrix of numeric features against categorical features.
# can use a larger sample since there are fewer plots and they're simpler in nature.

df_bike_samp = df_bike.sample(n=100, replace = False)


def boxgrid(x, y, **kwargs):
    """ Quick hack for creating box plots with seaborn's PairGrid. """
    default_color = sb.color_palette()[0]
    sb.boxplot(x=x, y=y, color=default_color)

plt.figure(figsize = [10, 10])
g = sb.PairGrid(data = df_bike_samp, y_vars = ['duration_sec', 'bike_id'], x_vars = categoric_vars,
                height = 3, aspect = 1.5)
g.map(boxgrid)
plt.show();
```

Interestingly, it appears that there are some negative relationships between the categorical variables and the two numeric variables of interest. The bike dataset with the best features (subscriber, female, no) seem to get the lowest duration seconds and bike_id, but also tend to be greatest on average.

```python
# since there's only three subplots to create, using the full data should be fine.
plt.figure(figsize = [8, 8])

# subplot 1: user_type vs member_gender
plt.subplot(3, 1, 1)
sb.countplot(data = df_bike, x = 'user_type', hue = 'member_gender', palette = 'Blues')

# subplot 2: bike_share_for_all_trip vs. member_gender
ax = plt.subplot(3, 1, 2)
sb.countplot(data = df_bike, x = 'bike_share_for_all_trip', hue = 'member_gender', palette = 'Blues')
ax.legend(ncol = 2) # re-arrange legend to reduce overlapping

# subplot 3: bike_share_for_all_trip vs. user_type, use different color palette
ax = plt.subplot(3, 1, 3)
sb.countplot(data = df_bike, x = 'bike_share_for_all_trip', hue = 'user_type', palette = 'Greens')
ax.legend(loc = 1, ncol = 2) # re-arrange legend to remove overlapping

plt.show()
```

There doesn't seem to be that much interaction between df_bike user_id and member_gender, though proportionally it seems like there might be more subscriber *user_type* on the greatest *member_gender* (*male and female*). There are no interactions on *the bike_share_for_all_trip* and *gender_member*, the lower level for both *yes* and *no* is the *other*. It looks like there are more *male* in the member_gender.

```python
# scatter plot of duration_sec vs. bike_id, with log transform on price axis

plt.figure(figsize = [8, 6])
plt.scatter(data = df_bike, x = 'bike_id', y = 'duration_sec')
plt.xlabel('bike_id')
plt.ylabel('duration_sec')
```

```python
def cuberoot_trans(x, inverse = False):
    """ quick function for computing cube root and cube operations """
    if not inverse:
        return x ** (1/3)
    else:
        return x ** 3
```

```python
# compute the logarithm of duration_sec to make multivariate plotting easier
def log_trans(x, inverse = False):
    """ quick function for computing log and power operations """
    if not inverse:
        return np.log10(x)
    else:
        return np.power(10, x)

df_bike['lod_duration_sec'] = df_bike['duration_sec'].apply(log_trans)
df_bike['cr_bike_id'] = df_bike['bike_id'].apply(cuberoot_trans)
```

```python
# plot the categorical variables against duration_sec and bike_id again, this time
# with full data and variable transforms
bike_id_ticks = [2, 3, 5, 7, 9, 10, 15, 20]
fig, ax = plt.subplots(ncols = 2, nrows = 3 , figsize = [10,10])

for i in range(len(categoric_vars)):
    var = categoric_vars[i]
    sb.violinplot(data = df_bike, x = var, y = 'lod_duration_sec', ax = ax[i,0],
               color = def_color)
    ax[i,0].set_yticks(log_trans(np.array([500, 1e3, 2e3, 5e3, 1e4, 2e4])))
    ax[i,0].set_yticklabels([500, '1k', '2k', '5k', '10k', '20k'])
    sb.violinplot(data = df_bike, x = var, y = 'cr_bike_id', ax = ax[i,1],
               color = def_color)

plt.show()
```

The plot of the full data using a violin plot and variable transformations reveals much more than the earlier box plots. It is highly suggested that the decrease in average duration second across increasing quality is due to the average quality decreasing with increasing quality. This is clearest in the plots across member_gender, moderately visible with user_type and bike_share_for_all_trip.

```python
# Use group_by() and size() to get the number of member_gender and each combination of the two variable
ct_counts = df_bike.groupby(['member_gender', 'bike_share_for_all_trip']).size()
ct_counts
```

```python
ct_counts = ct_counts.reset_index(name='count')
ct_counts
```

```python
# Use DataFrame.pivot() to rearrange the data, to have vehicle class on rows
ct_counts = ct_counts.pivot(index = 'member_gender', columns = 'bike_share_for_all_trip', values = 'count')
ct_counts
```

```python
sb.heatmap(ct_counts)
```

```python
sb.heatmap(ct_counts, annot = True, fmt = 'd')
```

The heat map show the relationship between bike_share_for_all_trip and member_gender, you can see the relationships as male having the highest section.

```python
gen_vars = ['Female', 'Male','Other']
vclasses = pd.api.types.CategoricalDtype(ordered=True, categories=gen_vars)
df_bike['member_gender'] = df_bike['member_gender'].astype(vclasses);

# Plot the Seaborn's FacetGrid
g = sb.FacetGrid(data = df_bike, col = 'member_gender')
g.map(plt.hist, "bike_id")
```

```python
sb.pointplot(data=df_bike, x='member_gender', y='duration_sec', color=def_color, errorbar='sd', linestyles="")
plt.xticks(rotation=15);
plt.ylabel('duration_sec')
```

```python
df_bike_flag = (df_bike['duration_sec'] >= 100) & (df_bike['duration_sec'] <= 1000)
df_bike_sml = df_bike.loc[df_bike_flag,:]
```

```python
fig = plt.figure(figsize = [8,6])
ax = sb.pointplot(data = df_bike_sml, x = 'member_gender', y = 'duration_sec', hue = 'bike_share_for_all_trip',
           palette = 'Blues', linestyles = '', dodge = 0.4)
plt.title('')
plt.ylabel('duration_sec')
plt.show();
```
